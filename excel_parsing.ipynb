```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Time'] = df['Timestamp'].dt.time
df['Date'] = df['Timestamp'].dt.date
df.drop('Timestamp', axis=1, inplace=True)
df
```

```python
# @title Gender vs How much Water  you in-take per day(in liters)?

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['How much Water  you in-take per day(in liters)?'].value_counts()
    for x_label, grp in df.groupby('Gender')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Gender')
_ = plt.ylabel('How much Water  you in-take per day(in liters)?')
```

```python
# @title Distribution of Sleep Hours

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.pie(df['How many hours of sleep do  you get per night(on an average)?'].value_counts(), labels=df['How many hours of sleep do  you get per night(on an average)?'].value_counts().index, autopct='%1.1f%%')
_ = plt.title('Distribution of Sleep Hours')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(df['Weigh(in kgs)'], df['height (in m)'], color='skyblue')
plt.title('Height and Weight')
plt.xlabel('Weight (in kgs)')
plt.ylabel('Height (in m)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['BMI'] = scaler.fit_transform(df['BMI(weight/height^2)'].values.reshape(-1,1))
df['BMI']
```

```python
from matplotlib import pyplot as plt
df['height (in m)'].plot(kind='line', figsize=(8, 4), title='height (in m)')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()
```
